```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
data = pd.read_csv('flux_drop.csv', delimiter=', ', engine='python')

# Extract the columns you want to plot
x = data['t(in s)']
f = data['flux (mol/ms)']
m = data['n_e_l(t)(mol)']
c = data['c_e_g(t)(mol/m3)']


fig, ax = plt.subplots(1,1,dpi =150)
axs = [ax]

# Create a line plot
axs[0].plot(x, m, marker='', ls='-', color='blue') 

# Add labels and title
# plt.ylim(0,0.8E-05)
axs[0].set_xlabel('time (in seconds)')
axs[0].set_ylabel('n_e_l(t), $mol$')

axs[0].grid()

fig.savefig('n_e_l.pdf', bbox_inches='tight')
# Display the plot
# plt.show()
```

```python
fig, ax = plt.subplots(1,1,dpi =150)
axs = [ax]

# Create a line plot
axs[0].plot(x, c, marker='', linestyle='-', color='blue')  # Adjust marker, linestyle, and color as neededax[0]
#ax[0]dd labels and title
axs[0].set_ylim(0.0,1.1)
#axs[0].set_xlim(0.0,0.16)
axs[0].set_xlabel('time (in seconds)')
axs[0].set_ylabel('c_e_g(t), $mol/m^3$')
#ax[0]t.title('flux')
axs[0].grid()

fig.savefig('c_e_g.pdf', bbox_inches='tight' )
# Display the plot
# plt.show()
```
