```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Read the CSV file
data = pd.read_csv('Gill2008_extended.csv', delimiter=', ', engine='python')

# Extract the columns you want to plot
x = data['x']
y = data['y']

h = y.values/x.values

# Fit a second-degree polynomial (line) to the data
coefficients4 = np.polyfit(x, y, 4)
poly = np.poly1d(coefficients4)

# Print the equation of the fitted polynomial
equation = poly.__str__()
print('Fitted Polynomial Equation:', equation)

# Extract the coefficients
a, b, c ,d ,e  = coefficients4  # Quadratic coefficients: a*x^2 + b*x + c

print("4th order poly Coefficients:")
print("a:", a)
print("b:", b)
print("c:", c)
print("d:", d)
print("e:", e)

x_e_l = 0.11133044049865734
test = poly(x_e_l)
print("test: " +str(test))
print("H: " +str(test/0.2571018575143514))
fig, ax = plt.subplots(1,1,dpi =150)
axs = [ax]

# Create a line plot
axs[0].plot(x, y, marker='.', ls='-', color='blue')  # Adjust marker, linestyle, and color as needed
axs[0].plot(x, poly(x), 'r-', label='Fitted Polynomial')

# Add labels and title
plt.ylim(-0.01,0.9)
axs[0].set_xlabel('mole fraction of ethanol in liquid')
axs[0].set_ylabel('mole fraction of ethanol in vapor')
plt.title('data extracted from Gill2008, Fig 1a')
axs[0].grid()
```

```python
# The value you want to achieve after multiplication
target_value = 0.936214

# Calculate the values to multiply each original value by
alpha = [target_value / value for value in h]
del alpha[0]
h = h[1:]
# Fit a second-degree polynomial (line) to the data
coeffs4 = np.polyfit(h, alpha, 4)
poly = np.poly1d(coeffs4)

# Print the equation of the fitted polynomial
equation = poly.__str__()
print('Fitted Polynomial Equation:', equation)

# Extract the coefficients
a, b, c ,d ,e  = coeffs4  # Quadratic coefficients: a*x^2 + b*x + c

print("4th order poly Coefficients:")
print("a:", a)
print("b:", b)
print("c:", c)
print("d:", d)
print("e:", e)

fig, axs1 = plt.subplots(1,1,dpi =150)
axs1.plot(h, alpha, marker='.', ls='-', color='blue')  # Adjust marker, linestyle, and color as needed
axs1.plot(h, poly(h), 'r-', label='Fitted Polynomial')
#plt.ylim(-0.01,0.9)
axs1.set_xlabel('H from Gill')
axs1.set_ylabel('alpha values')
plt.title('alpha values to get H x alpha = 0.936')
axs1.grid()
```
