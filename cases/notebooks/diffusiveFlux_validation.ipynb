```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.ticker import ScalarFormatter
from natsort import natsorted

import pandas as pd
import matplotlib.tri as mtri
import csv
```

```python
# Read the experimental flux CSV file into a DataFrame
df2 = pd.read_csv('../../data/expFlux_mol_per_s.csv')
expFlux = df2['flux(mol/s)'].values

data3 = pd.read_csv('../convergence_study/mesh3/flux_drop.csv', delimiter=', ', engine='python') ### mesh3 
x3 = data3['t(in s)']
f3 = data3['flux (mol/ms)']

fig, ax = plt.subplots(1,1,dpi =150)
axs = [ax]

df = pd.read_csv('../convergence_study/mesh3/0.960C_mesh3.csv')#, delimiter=', ', engine='python')

flux = df['flux(mol/s)'].values
time = df['time(s)'].values

axs[0].scatter(time,flux,s=5, color = 'green', label='$c_{\mathrm{E}}^{\mathrm{G}}(0)|_{\Sigma} = 0.960 / \mathrm{(mol/m^3)}$')
axs[0].scatter(time,expFlux,s=5, color = 'grey', label='experimental flux')

flux_in_mol_s_3 = [i * 0.060 for i in f3]
axs[0].plot(x3[:24000], flux_in_mol_s_3[:24000], marker='', ls='-', color='green', label='$c_{\mathrm{E}}^{\mathrm{G}}(0)|_{\Sigma} = 0.960 / \mathrm{(mol/m^3)}$')

formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((0, -5)) 
axs[0].xaxis.set_major_formatter(formatter)
axs[0].yaxis.set_major_formatter(formatter)
axs[0].legend()
axs[0].set_xlim(right=25)
axs[0].set_ylim(0.0E-06,1.1E-06)
axs[0].legend(loc='lower right',prop={'size': 8})#(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=4) #loc=2, prop={'size': 7}
axs[0].set_xlabel(xlabel="time / (seconds)", labelpad=0.4)
axs[0].set(ylabel='flux / (mol/s)')
#ax1.text(0.5, -0.32, 'b)', horizontalalignment='center', verticalalignment='center', transform=ax2.transAxes)
axs[0].grid()#(b=None, which='major', axis='both', linestyle='dotted', lw=0.5)
#ax2.set_title('9 wt percent ethanol loss', fontsize=8)
fig.savefig('../figures/diffFluxValidation.pdf', bbox_inches='tight', dpi=600.0)
#fig.savefig('N_All_Profiles.pgf', bbox_inches='tight', dpi=600)
plt.show()
print('Done')
```
